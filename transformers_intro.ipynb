```python
from transformers import pipeline
classifier=pipeline('sentiment-analysis')
```

```python
classifier("I am Happy.")
```

```python
results=classifier(["i am happy","i am not happy"])
for result in results:
  print(result)
```

```python
from transformers import AutoTokenizer, TFAutoModelForSequenceClassification
```

```python
model_name="nlptown/bert-base-multilingual-uncased-sentiment"
model=TFAutoModelForSequenceClassification.from_pretrained(model_name,from_pt=True)
tokenizer=AutoTokenizer.from_pretrained(model_name)
classifier=pipeline("sentiment-analysis",model=model,tokenizer=tokenizer)
```

```python
classifier("i am not happy")
```

```python
input=tokenizer("i am not happy")
print(input)
```

```python
tf_batch=tokenizer(["i am happy.","i am not happy."],
                   padding=True,#Ensures that all input sequences are the same length by adding special padding tokens (usually zeros) to shorter sequences
                   truncation=True,# If any sentence is longer than the specified max_length, it will be truncated to fit.
                   max_length=512,#maximum length for truncation
                   return_tensors="tf")#"tf" indicates that the output should be in TensorFlow tensor format.
```

```python
for key,value in tf_batch.items():
  print(f"{key}:{value.numpy().tolist()}")
```
