 **Whitespace Tokenizer**

```python
from nltk.tokenize import WhitespaceTokenizer
text="good morning to you all, i hope everyone is happy and doing well."
tokenizer=WhitespaceTokenizer()
tokens=tokenizer.tokenize(text)
print(tokens)
```

**Word Tokenizer**

```python
import nltk
from nltk.tokenize import word_tokenize
nltk.download('punkt')
tokens=word_tokenize(text)
print(tokens)
```

**Sentence Tokenizer**

```python
from nltk.tokenize import sent_tokenize
text1="hello everyone. nice to meet you all."
sentences=sent_tokenize(text1)
print(sentences)
```

**Regex Tokenizer**

```python
from nltk.tokenize import RegexpTokenizer
tokenizer=RegexpTokenizer(r'\w+')
text1="good morning everyone, let's talk about some topic"
tokens=tokenizer.tokenize(text1)
print(tokens)
```

**Subword Tokenizer**

```python
from transformers import BertTokenizer
tokenizer=BertTokenizer.from_pretrained('bert-base-uncased')
text1="The BERT tokenizer is designed to tokenize text in a way that is compatible with BERT (Bidirectional Encoder Representations from Transformers). BERT tokenizers use a method called WordPiece, which breaks down words into smaller subwords or word pieces, allowing the model to handle rare words and out-of-vocabulary terms more effectively."
tokens=tokenizer.tokenize(text1)
print(tokens)
```

**Byte-Pair Tokenizer**

```python
from transformers import GPT2Tokenizer
tokenizer=GPT2Tokenizer.from_pretrained("gpt2")
tokens=tokenizer.tokenize(text)
print(tokens)
```

**XLNetTokenizer**

```python
from transformers import XLNetTokenizer
tokenizer = XLNetTokenizer.from_pretrained("xlnet/xlnet-base-cased")
tokens=tokenizer.tokenize("Don't you just love Transformers? We sure do.")
print(tokens)
```
